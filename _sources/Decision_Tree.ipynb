Sekarang kita akan sesuaikan perhitungan manual decision tree berdasarkan data *Citrullus lanatus* dan *Citrullus amarus* dengan fitur **Diameter Buah (cm)**.

### **Ambil Threshold Split (contoh):**

Kita ambil threshold misalnya **20.5 cm**

* Group Left (≤ 20.5)
* Group Right (> 20.5)

---

### **Hitung Entropy Awal (sebelum split):**

Jumlah kelas:

* Citrullus lanatus: 8
* Citrullus amarus: 7

$$
p_1 = \frac{8}{15}, \quad p_2 = \frac{7}{15}
$$

$$
Entropy(D) = -\left(\frac{8}{15}\log_2\frac{8}{15} + \frac{7}{15}\log_2\frac{7}{15}\right)
$$

Hitung:

$$
\frac{8}{15} \approx 0.5333, \quad \frac{7}{15} \approx 0.4667
$$

$$
Entropy(D) \approx -0.5333 \cdot \log_2 0.5333 - 0.4667 \cdot \log_2 0.4667
$$

$$
\approx -0.5333 \cdot (-0.9069) - 0.4667 \cdot (-1.0984)
$$

$$
\approx 0.4836 + 0.5122 = 0.9958 \approx 0.996
$$

---

### **Entropy Setelah Split**

#### Group Left (≤ 20.5):

* Data: No 9, 6, 7, 15, 13, 8, 4, 10 → 8 data
* Kelas:

  * Citrullus amarus = 7
  * Citrullus lanatus = 1

$$
p_{amarus} = \frac{7}{8}, \quad p_{lanatus} = \frac{1}{8}
$$

$$
Entropy(left) = -\left( \frac{7}{8}\log_2 \frac{7}{8} + \frac{1}{8}\log_2 \frac{1}{8} \right)
$$

$$
\frac{7}{8} = 0.875, \quad \frac{1}{8} = 0.125
$$

$$
Entropy(left) \approx -0.875 \cdot (-0.1926) - 0.125 \cdot (-3)
$$

$$
\approx 0.1685 + 0.375 = 0.5435 \approx 0.543
$$

#### Group Right (> 20.5):

* Data: No 14, 2, 1, 12, 5, 3, 11 → 7 data
* Kelas:

  * Citrullus lanatus = 7

$$
Entropy(right) = 0
$$

---

### **Information Gain (IG):**

$$
IG = Entropy(D) - \left( \frac{8}{15} \cdot Entropy(left) + \frac{7}{15} \cdot Entropy(right) \right)
$$

$$
IG = 0.996 - \left( \frac{8}{15} \cdot 0.543 + \frac{7}{15} \cdot 0 \right)
$$

$$
= 0.996 - \left( 0.5333 \cdot 0.543 \right)
$$

$$
= 0.996 - 0.2896 = 0.7064 \approx \boxed{0.706}
$$

---

### **Gini Index**

#### Gini Awal:

$$
Gini(D) = 1 - \left( \left(\frac{8}{15}\right)^2 + \left(\frac{7}{15}\right)^2 \right)
$$

$$
= 1 - \left( 0.2844 + 0.2178 \right) = 1 - 0.5022 = 0.4978 \approx 0.497
$$

#### Gini Left:

$$
Gini(left) = 1 - \left( \left( \frac{7}{8} \right)^2 + \left( \frac{1}{8} \right)^2 \right)
$$

$$
= 1 - (0.765625 + 0.015625) = 1 - 0.78125 = 0.21875
$$

#### Gini Right:

$$
Gini(right) = 0 \quad \text{(pure class)}
$$

#### Gini Split:

$$
Gini_{split} = \frac{8}{15} \cdot 0.21875 + \frac{7}{15} \cdot 0
$$

$$
= 0.5333 \cdot 0.21875 = 0.1166 \approx \boxed{0.116}
$$

---

### **Hasil Akhir untuk fitur Diameter Buah (cm):**

| Ukuran Evaluasi  | Nilai |
| ---------------- | ----- |
| Entropy Awal     | 0.996 |
| Information Gain | 0.706 |
| Gini Split       | 0.116 |

## Visualisasi Decision Tree

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# Data
data = {
    "Diameter Buah (cm)": [22.0, 21.5, 23.0, 20.5, 22.5, 18.5, 19.0, 20.0, 18.0, 20.5, 23.5, 22.0, 19.5, 21.0, 19.0],
    "Berat Buah (kg)": [6.5, 6.0, 7.0, 5.5, 6.8, 4.0, 4.2, 5.0, 3.8, 5.2, 7.5, 6.3, 4.5, 6.0, 4.3],
    "Kelas": [
        "Citrullus lanatus", "Citrullus lanatus", "Citrullus lanatus", "Citrullus lanatus", "Citrullus lanatus",
        "Citrullus amarus", "Citrullus amarus", "Citrullus amarus", "Citrullus amarus", "Citrullus amarus",
        "Citrullus lanatus", "Citrullus lanatus", "Citrullus amarus", "Citrullus lanatus", "Citrullus amarus"
    ]
}
df = pd.DataFrame(data)

# Label Encoding
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # lanatus=1, amarus=0 (otomatis tergantung le)

# Fitur dan Target
X = df[["Diameter Buah (cm)", "Berat Buah (kg)"]]
y = df["Label"]

# Train Decision Tree
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# Tampilkan Pohon Keputusan
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)
```

Kode diatas melakukan klasifikasi jenis buah semangka menggunakan algoritma Decision Tree. Data diameter dan berat buah dimasukkan dalam bentuk DataFrame. Label kelas buah yang berupa teks diubah menjadi angka menggunakan LabelEncoder. Fitur yang digunakan adalah diameter dan berat buah, sedangkan targetnya adalah label hasil encoding. Model Decision Tree dilatih menggunakan DecisionTreeClassifier dengan kriteria gini dan maksimal kedalaman pohon 3. Setelah model dilatih, struktur pohon keputusan ditampilkan dalam bentuk teks menggunakan export_text().

## Visualisasi Splitting

```python
# Uji 2 Data Baru
uji_baru = pd.DataFrame({
    'Diameter Buah (cm)': [19.0, 22.8],
    'Berat Buah (kg)': [4.5, 6.9]
})

prediksi = clf.predict(uji_baru)
hasil = le.inverse_transform(prediksi)

print("\n=== Prediksi 2 Data Baru ===")
for i, (d, b) in enumerate(zip(uji_baru['Diameter Buah (cm)'], uji_baru['Berat Buah (kg)'])):
    print(f"Data {i+1}: Diameter={d}, Berat={b} → Prediksi: {hasil[i]}")

# Visualisasi Decision Boundary
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"
plot_step = 0.1  # lebih kecil biar halus

x_min, x_max = X['Diameter Buah (cm)'].min() - 1, X['Diameter Buah (cm)'].max() + 1
y_min, y_max = X['Berat Buah (kg)'].min() - 1, X['Berat Buah (kg)'].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Diameter Buah (cm)', 'Berat Buah (kg)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Diameter Buah (cm)'], X.iloc[idx]['Berat Buah (kg)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

plt.xlabel("Diameter Buah (cm)")
plt.ylabel("Berat Buah (kg)")
plt.title("Decision Boundary (Diameter vs Berat)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Kode ini menguji dua data baru dengan model Decision Tree yang telah dilatih, lalu menampilkan hasil prediksinya. Setelah itu, kode membuat visualisasi decision boundary untuk menunjukkan pembagian kelas berdasarkan diameter dan berat buah. Area keputusan digambar dengan contourf(), sedangkan data latih ditampilkan sebagai scatter plot.
